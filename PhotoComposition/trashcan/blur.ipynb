```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#画像を読み込む関数
def load_img():
    #0-1にするために255で割る
    img = cv2.imread('image/background_2.jpg').astype(np.float32) / 255
    #カラーをBGRからRGBに変換(そのままでは色が変になる)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
#画像を表示する関数
def display_img(img):
    #画像の縦横の大きさを指定
    fig = plt.figure(figsize=(12,10))
    #表示の位置を指定(111は画像の左上を意味)(引数は行，列，位置を示している)
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
#画像とテキストを合成
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text='OpenCV',org=(50,250), fontFace=font,fontScale= 4,color=(255,0,0),thickness=2)
display_img(img)
```

```python
#h平均化
blurred_img = cv2.blur(img,ksize=(5,5))
display_img(blurred_img)
```

```python
#ガウシアンフィルタ
blurred_img = cv2.GaussianBlur(img,(5,5),10)
display_img(blurred_img)
```

```python
#バイラテラルフィルタ
blur = cv2.bilateralFilter(img,9,75,75)
display_img(blur)
```
