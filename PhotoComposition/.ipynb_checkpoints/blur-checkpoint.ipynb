```python
import cv2
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
#ファイル読み込み
tzuyu = cv2.imread('./tzuyu/0001.png', -1)# -1を付けることでアルファチャンネルも読んでくれるらしい。
background = cv2.imread('./landscape/0001.jpg')

#サイズ確認
print("tzuyu"+str(tzuyu.shape))
print("background"+str(background.shape))

#指定の倍率に画像を縮小
# 高さ固定のリサイズ関数
def scale_to_height(img,height):
    scale = height / img.shape[0]
    return cv2.resize(img,dsize=None,fx=scale,fy=scale)
tzuyu =scale_to_height(tzuyu,300)
background=scale_to_height(background,400)

#backgroundが横長になるので，縦長にトリミング
b_height,b_width = background.shape[:2]
background=background[0:b_height,b_width//2-200:b_width//2+200]

#サイズ確認
print("tzuyu"+str(tzuyu.shape))
print("background"+str(background.shape))

#backgroundにガウシアンフィルタ((15,15),0:この数字を大きくしたらボケが強くなる)
background = cv2.GaussianBlur(background,(15,15),0)
plt.imshow(tzuyu)
plt.imshow(background)
```

```python
#マスク画像作成
mask = tzuyu[:,:,3]  # アルファチャンネルだけ抜き出す
mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)  # 3色分に増やす
mask = mask / 255.0  # 0-255だと使い勝手が悪いので、0.0-1.0に変更
tzuyu = tzuyu[:,:,:3]  # アルファチャンネルを削除
mask=mask.astype(np.uint8)#mask,tzuyu,backgroundのデータ型を統一

"""
print("mask:"+str(mask.dtype)+str(mask.shape))
print("tzuyu:"+str(tzuyu.dtype)+str(tzuyu.shape))
print("background:"+str(background.dtype)+str(background.shape))

"""
#tzuyuを表示する位置を指定(backgroundの中央下部から出る感じで)                        
t_height,t_width = tzuyu.shape[:2]
b_height,b_width = background.shape[:2]
background[b_height-t_height:b_height,b_width//2-t_width//2:b_width//2+t_width//2] *= 1 - mask  # 透過率に応じて元の画像を暗くする。
background[b_height-t_height:b_height,b_width//2-t_width//2:b_width//2+t_width//2] += tzuyu * mask  # 貼り付ける方の画像に透過率をかけて加算。

#backgroundにガウシアンフィルタ（2枚の画像の境界に違和感があるため，全体を少しだけぼかす）
background = cv2.GaussianBlur(background,(3,3),10)

#保存
cv2.imwrite('output/output.jpg', background)
img=Image.open('output/output.jpg')
img
```
