```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Read images : src image will be cloned into dst
#im = cv2.imread("image/wood-texture.jpg")
#im = cv2.imread("image/sky_1.jpg")
im = cv2.imread("image/lab.jpg")
#obj= cv2.imread("image/iloveyouticket.jpg")
#obj= cv2.imread("image/airplane.jpg")
obj= cv2.imread("image/tzuyu_12.png")
#obj= cv2.imread("image/pikachu.png")

# Create an all white mask
#マスクは白で二つの画像のテクスチャを混ぜるので，貼り付けたいものの彩度が高く，背景が低ければ違和感なく貼り付けられる
mask = 255 * np.ones(obj.shape, obj.dtype)
plt.imshow(mask)
```

```python
# The location of the center of the src in the dst
width, height, channels = im.shape
center = (int(height/2), int(width/2))

# Seamlessly clone src into dst and put the results in output
normal_clone = cv2.seamlessClone(obj, im, mask, center, cv2.NORMAL_CLONE)
mixed_clone = cv2.seamlessClone(obj, im, mask, center, cv2.MIXED_CLONE)
monochrome_clone = cv2.seamlessClone(obj, im, mask, center, cv2.MONOCHROME_TRANSFER)
#plt.imshow(mixed_clone)
# Write results
cv2.imwrite("output/opencv-normal-clone-example.jpg", normal_clone)
cv2.imwrite("output/opencv-mixed-clone-example.jpg", mixed_clone)
cv2.imwrite("output/opencv-monochrome-clone-example.jpg", monochrome_clone)
img1=Image.open('output/opencv-normal-clone-example.jpg')
img2=Image.open('output/opencv-mixed-clone-example.jpg')
img1
```
